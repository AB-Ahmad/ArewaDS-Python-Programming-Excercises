```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
#genrating a simple random interger
random_int = np.random.randint(1,101,size= 12)
print(random_int)
```

```python
#generate a random float number 
random_float = np.random.uniform(0,1,size = 12)
print(random_float)

```

```python
np.random.seed(42)
#random seeds for reproducibility
age = np.random.randint(15,50, size = 100)
class1 = np.random.randint(1,5,size = 100)
grade = np.random.uniform(0,5,size =100)
print(age)
print(class1)
print(grade)
```

```python
#dataframe
my_data = pd.DataFrame({"Student Age":age, "Level":class1,"grade":grade})
my_data.head(10)
```

```python
plt.figure(figsize= (10,6))
plt.title("Students info")
plt.xlabel(age,)
plt.ylabel(class1)
plt.tight_layout()
plt.show()
```
